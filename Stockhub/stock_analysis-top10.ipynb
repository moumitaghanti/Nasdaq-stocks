```python
import pandas as pd
from pandas.plotting import scatter_matrix
import datetime
import pandas_datareader.data as web
from pandas import Series, DataFrame


start = datetime.datetime(2014, 1, 1)
end = datetime.datetime(2019, 12, 31)

ticker = ["AAPL","MSFT","AMZN","FB","JPM","XOM","GOOG", "GOOGL","JNJ"]
          #,"BRK.B","JPM","XOM","GOOG", "GOOGL","JNJ"]
appendDf =[]
for tick in ticker:
    df = web.DataReader(tick, 'yahoo', start, end)
    df["Ticker"] = tick
    appendDf.append(df)
    
appendDf = pd.concat(appendDf)

appendDf.reset_index()
appendDf.tail()
```

```python
stocksDf = appendDf.rename(columns={"Date":"date","High":"high","Low":"low",
                                    "Open":"open","Close":"close","Volume":"volume",
                                    "Adj Close":"adj_close", "Ticker":"ticker"})
```
